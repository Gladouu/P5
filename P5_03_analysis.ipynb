# Clustering

## Clustering Hiérarchique Ascendant

> *Le clustering permet de regrouper des individus similaires, c'est-à-dire qu'il va partitionner l'ensemble des individus. On cherche donc à ce que les groupes soient :*

> * *Resserrés sur eux-mêmes : deux points qui sont proches devraient appartenir au même groupe*
> * *Loin les uns des autres, c’est-à-dire qu’ils soient fortement différenciés*

```python
#Preparing data for clustering
df = df.set_index('Pays')
X = df.values
names = df.index

#Centering and reduction
std_scale = preprocessing.StandardScaler().fit(X)
X_scaled = std_scale.transform(X)

# Hierarchical Clustering 
Z = linkage(X_scaled, 'ward')

#dendrogram
plot_dendrogram(Z, names, 5)
```

```python
#form clusters
clusters = fcluster(Z, 5, criterion='maxclust')
df["Cluster"] = clusters

#export
df.sort_values('Cluster').to_csv('Liste des pays par cluster.csv', index=True)
```

```python
# ------- PART 1: Define a function that do a plot for one line of the dataset
 
def make_spider(row, title, color):

    #number of variable
    categories=list(XCR)[1:]
    N = len(categories)

    #What will be the angle of each axis in the plot? (we divide the plot / number of variable)
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]

    #Initialise the spider plot
    ax = plt.subplot(5,2,row+1, polar=True)
    plt.subplots_adjust(hspace=0.6)

    #If you want the first axis to be on top:
    ax.set_theta_offset(pi / 2)
    ax.set_theta_direction(-1)

    #Draw one axe per variable + add labels labels yet
    plt.xticks(angles[:-1], categories, color='grey', size=10)
    
    #Go through labels and adjust alignment based on where it is in the circle.
    for label, angle in zip(ax.get_xticklabels(), angles):
        if angle in (0, np.pi) :
            label.set_horizontalalignment('center')
        elif 0 < angle < np.pi :
            label.set_horizontalalignment('left')       
        else :
            label.set_horizontalalignment('right')
    
    # Draw ylabels
    ax.set_rlabel_position(0)
    plt.yticks([-2,-1,0,1,2], color="grey", size=7)
    plt.ylim(-3, 3)

    # Ind1
    values=XCR.loc[row].drop('Cluster').values.flatten().tolist()
    values += values[:1]
    ax.plot(angles, values, color=color, linewidth=2, linestyle='solid')
    ax.fill(angles, values, color=color, alpha=0.4)

    # Add a title
    plt.title(title, size=17, color=color, y=1.1, pad=20)

    
# ------- PART 2: Apply the function to all individuals
# initialize the figure
my_dpi=100
plt.figure(figsize=(1800/my_dpi, 1800/my_dpi), dpi=my_dpi)
 
# Create a color palette:
my_palette = plt.cm.get_cmap("Set2", len(XCR.index))
 
# Loop to plot
for row in range(0, len(XCR.index)):
    make_spider(row=row, title='Cluster '+ str(XCR['Cluster'][row]), color=my_palette(row))
    
#save
plt.savefig('Radar Chart.jpg', dpi=500, bbox_inches='tight', pad_inches=0.5)
```

### Éboulis des valeurs propres

>*Finalement, combien de composantes analyser ?
Il n’y a pas de réponse toute faite : tout dépend de vos objectifs et du contexte. Tout dépend jusqu’où vous souhaitez pousser votre analyse.
Cependant, voici quelques pistes :*

> - *Tout d’abord, on a tendance à ne pas considérer comme importants les axes dont l’inertie associée est inférieure à (100/p)%, car ils représentent moins de variabilité qu’une seule variable initiale. La valeur de (100/p)% est celle obtenue quand toutes les variables sont indépendantes deux à deux. Ce critère est appelé critère de Kaiser.*

> - *Ensuite, il existe la « méthode du coude » qui consiste à repérer l’endroit à partir duquel le pourcentage d’inertie diminue beaucoup plus lentement lorsque l’on parcourt le diagramme des éboulis de gauche à droite.*

> - *En pratique, on a plutôt tendance à s’arrêter dès que l’on n’arrive plus à interpréter les axes principaux d’inertie, c’est-à-dire quand on n’arrive plus à dire « le groupe de variables est très corrélé à Fs, et il correspond à la notion de [...] ». Il est même fréquent de n’analyser que le premier plan factoriel.*

### Contribution de chaque variable aux composantes principales

[Section 3.3.4 du document](http://eric.univ-lyon2.fr/~ricco/tanagra/fichiers/fr_Tanagra_ACP_Python.pdf)

```python
#number of observations 
n = df_pca.shape[0]

#number of variables
p = df_pca.shape[1]

#instantiation
pca = decomposition.PCA(svd_solver='full')

#calculations
coord = pca.fit_transform(X_scaled)

#corrected value
eigval = (n-1)/n*pca.explained_variance_ #or acp.singular_values_**2/n

#square root of eigenvalues
sqrt_eigval = np.sqrt(eigval)

#correlation of variables with axes
corvar = np.zeros((p,p))
for k in range(p):
    corvar[:,k] = pca.components_[k,:] * sqrt_eigval[k]

#cosine square of variables
cos2var = corvar**2

#contributions
ctrvar = cos2var
for k in range(p):
    ctrvar[:,k] = ctrvar[:,k]/eigval[k]
    
#show
pd.DataFrame({'variable':df_pca.columns,'F1':ctrvar[:,0],'F2':ctrvar[:,1], 'F3':ctrvar[:,2]})
```

## Test de Student

[Independent T-test](https://www.pythonfordatascience.org/independent-samples-t-test-python/)

>*The indepentent T-test is a parametric test used to test for a statistically significant difference in the means between 2 groups. As with all parametric tests, there are certain conditions that need to be met in order for the test results to be considered reliable.*

>*Parametric test assumptions*

>* *Population distributions are normal*
>* *Samples have equal variances*
>* *The two samples are independent*

### Hypothèse H0

$\left\{
\begin{array}{l} 
 H_0\ : \  μ = μ_0 \\
 H_1\ : \ μ ≠ μ_0
\end{array}
\right.$

## Assumption Check

### La distribution de la population suit une loi normale

$\left\{
\begin{array}{l} 
 H_0\ : \  \gamma = \gamma_0\ \ ,\ la\ variable\ suit\ donc\ une\ loi\ normale \\
 H_1\ : \ \gamma ≠ \gamma_0\ \ ,\ la\ variable\ ne\ suit\ pas\ une\ loi\ normale
\end{array}
\right.$

### Homoscédasticité 

$\left\{
\begin{array}{l} 
 H_0\ : \ les\ variances\ sont\ égales\  \\
 H_1\ : \ les\ variances\ ne\ sont\ pas\ égales\
\end{array}
\right.$

```python
#boxplot
color = sns.color_palette('pastel')

plt.figure(figsize=(30,15)) 

sns.boxplot(data=df[(df["Cluster"]== 4) | (df["Cluster"]== 5)], x='Dispo alim prot g/pers', y='Cluster', orient='h', palette=color,
            fliersize=4 , showfliers=True, showmeans=True, meanprops={"marker":"o", 
                                                                      "markerfacecolor":"white",
                                                                      "markeredgecolor":"white", 
                                                                      "markersize":"10"})

plt.xlabel('Dispo alim en protéine', fontsize=20, labelpad=30, fontweight='bold')
plt.xticks(fontsize=20)
plt.ylabel('Cluster', fontsize=20, labelpad=30, fontweight='bold')
plt.yticks(fontsize=20)
plt.title('Disponibilité alimentaire en protéine pour les cluster 4 et 5', fontsize=35, pad=50)

plt.show()
```

```python
pg.homoscedasticity(c4c5, dv='Dispo alim prot g/pers', 
                    group='Cluster', method='levene', 
                    alpha=0.05)
```

## Statistique du test

```python
pg.ttest(df['Dispo alim prot g/pers'][df['Cluster'] == 4],
         df['Dispo alim prot g/pers'][df['Cluster'] == 5],
         paired=False,
         tail='two-sided',
         confidence=0.95)
```

## Conclusion

```python
α = 0.05

if α > pg.ttest(df['Dispo alim prot g/pers'][df['Cluster'] == 4],
             df['Dispo alim prot g/pers'][df['Cluster'] == 5]).iloc[0,3] : 
    
    print("La p-value étant inférieure au risque α, on rejette donc H0, les moyennes des deux groupes sont différentes.")
else :
    print("La p-value étant supérieur au risque α, H0 est donc vrai, les moyennes des deux groupes sont égales.")
```
